## STATEWISE PROPERTY DETAILS 

```python
#Number Of bedrooms
bedcount = pd.DataFrame(df.groupby(['state'])['bed'].mean().reset_index())
bedcount.columns = ['State', 'Avg_Bedrooms']
bedcount['Avg_Bedrooms'] = bedcount['Avg_Bedrooms'].round(0)
#NUmber Of Properties
property_by_State = pd.DataFrame(df['state'].value_counts()).reset_index()
property_by_State.columns = ['State', 'Property Count']
result_df = pd.merge(property_by_State, bedcount[['State', 'Avg_Bedrooms']], on='State', how='left')

#Average Number of Bathrooms and Price
avg_df = df.groupby('state').agg({'bath': 'mean','price':'mean'}).reset_index()
avg_df.columns = ['State', 'Avg_bathrooms','Avg_Price']
result_df = pd.merge(result_df, avg_df, on='State', how='left')

#Number of cities
City_count = pd.DataFrame(df.groupby(['state'])['city'].nunique().reset_index())
City_count.columns = ['State','Citycount']
result_df = pd.merge(result_df, City_count, on='State', how='left')

#Property Per city
result_df['Property_per_city'] = result_df['Property Count']/result_df['Citycount']
result_df['Property_per_city'] = result_df['Property_per_city'].round(0)
result_df['Avg_bathrooms'] = result_df['Avg_bathrooms'].round(0)

#Status of property
status_count = pd.DataFrame(df.groupby(['state','status']).size().unstack(fill_value=0)).reset_index()
status_count.columns = ['State','For_Sale','Ready_to_Build']
result_df = pd.merge(result_df, status_count, on='State', how='left')
result_df.head(10)
```

## CITYWISE PROPERTY DETAILS 

```python
#CITY DETAILS
bedcount = pd.DataFrame(df.groupby(['city'])['bed'].mean().reset_index())
bedcount.columns = ['City', 'Avg_Bedrooms']
bedcount['Avg_Bedrooms'] = bedcount['Avg_Bedrooms'].round(0)

#Number of Prperties
property_by_City = pd.DataFrame(df['city'].value_counts()).reset_index()
property_by_City.columns = ['City', 'Property Count']
result_df = pd.merge(property_by_City, bedcount[['City', 'Avg_Bedrooms']], on='City', how='left')

#Number of Bathrooms
avg_df = df.groupby('city').agg({'bath': 'mean','price':'mean'}).reset_index()
avg_df.columns = ['City', 'Avg_bathrooms','Avg_Price']
result_df_city = pd.merge(result_df, avg_df, on='City', how='left')

#status Count
status_count = pd.DataFrame(df.groupby(['city','status']).size().unstack(fill_value=0)).reset_index()
status_count.columns = ['City','For_Sale','Ready_to_Build']
result_df_city = pd.merge(result_df_city, status_count, on='City', how='left')

#rund to decimal 0
result_df_city['Avg_bathrooms'] = result_df_city['Avg_bathrooms'].round(0) 

result_df_city.head(10)
```

### 1.Top Cities for you

This script allows you to filter real estate properties based on specific criteria and provides analysis for the top cities that meet the criteria.

User Inputs:
- Enter the maximum price: [User provides a numerical value]
- Enter the minimum number of bedrooms: [User provides a numerical value]
- Enter the minimum number of bathrooms: [User provides a numerical value]

The script then filters the dataset based on the user's criteria, calculates the property count per city, and presents the top cities based on the count.

Note: The filtering is done based on properties with a price less than or equal to the specified maximum, at least the specified minimum number of bedrooms, and at least the specified minimum number of bathrooms.

```python
def filter_real_estate(user_price, user_bed, user_bath):
    user_price = float(user_price)
    user_bed = float(user_bed)
    user_bath = float(user_bath)
    
    # Filter DataFrame based on user criteria
    filtered_df = df[
        (df['price'] <= user_price) &
        (df['bed'] >= user_bed) &
        (df['bath'] >= user_bath)
    ]
    
    # Calculate property count per city
    state_property_count = filtered_df.groupby('city').size().reset_index(name='count')
    
    # Sort by property count in descending order and get the top 5
    top_cities = state_property_count.sort_values(by='count', ascending=False).head(10)
    
    return top_cities

# User inputs
user_price_input = input("Enter the maximum price:")
user_bed_input = input("Enter the minimum number of beds:")
user_avg_bathroom_input = input("Enter the minimum average bathrooms:")

result = filter_real_estate(user_price_input, user_bed_input, user_avg_bathroom_input)
print('Top cities based on your criteria:')
print(result)
```

### 2. Property Analysis for a Selected City

This script enables you to analyze real estate properties in a selected city based on specific criteria.

User Inputs:
- Enter a city from the list below: [User selects a city from the available options]
- Enter the minimum number of bedrooms: [User provides a numerical value]
- Enter the minimum average bathrooms: [User provides a numerical value]

The script then filters the dataset for the selected city, considering the specified minimum number of bedrooms and average bathrooms. It calculates and presents statistics for the properties in the selected city, including the number of properties, average price, minimum price, and maximum price.

Note: The filtering is based on the selected city, a minimum number of bedrooms, and a minimum average number of bathrooms.

```python
unique_cities = pd.DataFrame(df['city'].sort_values(ascending =True).unique())
print("Available Cities:\n")
for city in unique_cities[0]:
    print(city)
```

```python
def filter_real_estate(selected_city, user_bedrooms, user_avg_bathrooms):
    user_bedrooms = float(user_bedrooms)
    user_avg_bathrooms = float(user_avg_bathrooms)
    filtered_df = df[
        (df['city'] == selected_city) &
        (df['bed'] >= user_bedrooms) &
        (df['bath'] >= user_avg_bathrooms)
    ]
    properties_count = filtered_df.shape[0]
    avg_price = filtered_df['price'].mean()
    min_price = filtered_df['price'].min()
    max_price = filtered_df['price'].max()
    return properties_count, avg_price, min_price, max_price

selected_city_input = input("Enter a city from the list above :")
user_bedrooms_input = input("Enter the minimum number of bedrooms: ")
user_avg_bathroom_input = input("Enter the minimum average bathrooms: ")

result = filter_real_estate(selected_city_input,user_bedrooms_input,user_avg_bathroom_input)
properties_count, avg_price, min_price, max_price = result #tuple to store the output

print(f"\nStatistics for {selected_city_input}:")
print(f"Number of Properties: {properties_count}")
print(f"Average Price: {avg_price:.2f}")
print(f"Minimum Price: {min_price:.2f}")
print(f"Maximum Price: {max_price:.2f}")
```
